```python
import matplotlib.pyplot as plt
plt.hist(selected_df[selected_df['SEXE'] == 1]['SEXE'], bins=[0.5, 1.5], align='mid', color='blue', label='Male', alpha=0.7)
plt.hist(selected_df[selected_df['SEXE'] == 2]['SEXE'], bins=[1.5, 2.5], align='mid', color='pink', label='Female', alpha=0.7)

plt.title('Histogram of SEXE Values')
plt.xlabel('SEXE')
plt.ylabel('Frequency')
plt.xticks([1, 2], ['Male', 'Female'])  # Adjust the labels accordingly
plt.legend()
plt.show()
```

```python
max_euro_by_sex = selected_df.groupby('SEXE')['EURO'].mean()

# Plot a bar plot to visualize the highest EURO values
max_euro_by_sex.plot(kind='bar', color=['blue', 'pink'])
plt.title('Highest EURO by SEXE')
plt.xlabel('SEXE')
plt.ylabel('EURO')
plt.xticks([0, 1], ['Male', 'Female'], rotation=0)
plt.show()
```

```python
average_euro_by_sex = selected_df.groupby('SEXE')['EURO'].mean()
average_euro_by_sex
```

```python
import seaborn as sns
heatmap_data = selected_df.pivot_table(index='CONT_TRAV', columns='SEXE', values='EURO', aggfunc='mean')

# Create a heatmap using seaborn
sns.heatmap(heatmap_data, cmap='YlGnBu', annot=True, fmt=".1f", cbar_kws={'label': 'Average EURO'})
plt.title('Heatmap based on CONT_TRAV')
plt.show()
```

```python
# Define age categories
bins = [20, 30, 40, 50, 60, 70]
labels = ['20-29', '30-39', '40-49', '50-59', '60-69']

# Create a new column for age categories
selected_df['AGE_CATEGORY'] = pd.cut(selected_df['AGE'], bins=bins, labels=labels, right=False)

# Group by age category and calculate the sum of EURO values
age_euro_sum = selected_df.groupby('AGE_CATEGORY')['EURO'].sum()

# Plot a pie chart
plt.pie(age_euro_sum, labels=age_euro_sum.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('Pie Chart based on Age and EURO')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
